# Crop images
This code goes through all images, crops, and saves.

```python
for j in range(0,len(files)):
    file = 'LicensePlates_testImages/' + files[j]
    bgr = cv2.imread(file)
    rgb = bgr[...,::-1]

    # Crop image
    height = np.size(rgb, 0)
    width = np.size(rgb, 1)
    wcent = int(round(width/2))
    hcent = int(round(height/2))
    cropped = rgb[hcent-200:hcent+200,wcent-300:wcent+300,:]
    
    #Save cropped image 400x600
    cropped = cropped[...,::-1]
    filename = "cropped_plates/img" + str(j) + ".jpg"
    cv2.imwrite( filename, cropped);
```
